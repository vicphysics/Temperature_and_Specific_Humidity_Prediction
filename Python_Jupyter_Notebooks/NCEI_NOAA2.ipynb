```python
import requests
from bs4 import BeautifulSoup
import re

# URL of the web page
url = 'https://www.ncei.noaa.gov/data/global-summary-of-the-day/access/2006'

# Send a GET request to the web page
response = requests.get(url)

# Parse the content of the page
soup = BeautifulSoup(response.content, 'html.parser')

# Find all links on the page
links = soup.find_all('a', href=True)

# Filter links that point to files (e.g. .csv)
file_links = [link['href'] for link in links if re.search(r'\\(.csv)$', link['href'], re.IGNORECASE)]

# Print the list of file links
for file_link in file_links:
   print(file_link)

print(len(file_links))
```

```python
import os
import requests
import pandas as pd
from io import StringIO

stations = []

latitude_min = 25.4
latitude_max = 49.4
longitude_min = -124.73
longitude_max = -66.95

# Loop through each file link
for file_link in file_links:
   url = 'https://www.ncei.noaa.gov/data/global-summary-of-the-day/access/2006/'+file_link
   response = requests.get(url)

   if response.status_code == 200:
      content = response.text
      content_df = pd.read_csv(StringIO(content))
      print(content_df.head())
      latitude = content_df['LATITUDE'].values[0]
      longitude = content_df['LONGITUDE'].values[0]
      elevation = content_df['ELEVATION'].values[0]
      #      if(latitude > latitude_min and latitude < latitude_max and longitude > longitude_min and longitude < longitude_max):
      print(latitude, longitude)
      print(content_df['STATION'].values[0])
      stations.append([content_df['STATION'].values[0], latitude, longitude, elevation])
   else:
      print(f"Failed to retrieve the file. Status code: {response.status_code}")

print(stations)
```

```python
df = pd.DataFrame(stations, columns = ['WMO_ID', 'latitude', 'longitude', 'elevation'])

# Write DataFrame to CSV
df.to_csv('WMO_Weather_Stations8.csv', index=False)
```
